```python
# Load OpenAI Gym and other necessary packages
import gym
import random
import numpy
import time
# Environment
env = gym.make("Taxi-v3")
env.reset()
env.render()


# Training parameters for Q learning
alpha = 0.9 # Learning rate
gamma = 0.9 # Future reward discount factor
num_of_episodes = 1000
num_of_steps = 500 # per each episode

# Q tables for rewards
#Q_reward = -100000*numpy.ones((500,6)) # All same
Q_reward = -100000*numpy.random.random((500, 6)) # Random

# Training w/ random sampling of actions
# YOU WRITE YOUR CODE HERE

obs = env.observation_space.n
act = env.action_space.n

epsilon = 1.0
q_table = numpy.zeros([obs, act])

for t in range(num_of_episodes):
    state = env.reset()
    step=0
    epochs=0
    penalty=0 
    reward=0
    done=False

    for step in range(num_of_steps):
    
        if random.uniform(0, 1) < epsilon:
            exAction = env.action_space.sample()    # Exploration
        else:
            exAction = numpy.argmax(q_table[state,:])     # Exploitation

        next_state, reward, done, info = env.step(exAction) 
        
        q_table[state, exAction] += alpha * (reward + gamma * numpy.max(q_table[next_state, :]) - q_table[state, exAction])

        state = next_state

        if reward == -10:
            penalty += 1

        if done == True:
            break


    epochs += 1
    t += 1
     


print("Training done.\n")


#print(q_table)

#Testing
#Could not get the original skeleton to perfectly work, so modified the testing part.

env.reset()
rewards = []
steps= []

for i in range(50):
  state = env.reset()
  step = 0
  done = False
  tot_rewards = 0
  tot_steps=0
  print('=====================')



  for step in range(num_of_steps):

    env.render()
    
    exAction = numpy.argmax(q_table[state, :])
    new_state, reward, done, info = env.step(exAction)

    step += tot_steps

    tot_rewards += reward

    if done:
      rewards.append(tot_rewards)
      print('Score: ', tot_rewards)

      steps.append(step)
      print('Steps: ', step)
      break

    state = new_state

env.close()

print("Average total reward:: {}".format(sum(rewards)/(50)))
print("Average number of actions: {}".format(sum(steps)/(50)))

```
